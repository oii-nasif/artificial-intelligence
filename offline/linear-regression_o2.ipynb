```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('https://raw.githubusercontent.com/oii-nasif/artificial-intelligence/master/datasets/linear-regression-train_o2.csv')
test_data = pd.read_csv('https://raw.githubusercontent.com/oii-nasif/artificial-intelligence/master/datasets/linear-regression-test_o2.csv')
```

```python
train_data = np.array(train_data)
test_data = np.array(test_data)

sz_train_data = len(train_data)
sz_test_data = len(test_data)
print(sz_train_data, sz_test_data)
```

```python
x_train = train_data[:, 0]
y_train = train_data[:, 1]

x_test = test_data[:, 0]
y_test = test_data[:, 1]
```

```python
x_train_mean = np.mean(x_train)
y_train_mean = np.mean(y_train)

print(x_train_mean)
print(y_train_mean)
```

```python
Sumxy = 0
Sumxx = 0
for i in range(sz_train_data):
    Sumxy += y_train[i]*x_train[i]
    Sumxx += x_train[i]**2.0
SSxy = Sumxy - sz_train_data*x_train_mean*y_train_mean
SSxx = Sumxx - sz_train_data*(x_train_mean**2)
print(SSxy)
print(SSxx)
```

```python
beta_1 = SSxy/SSxx
beta_0 = y_train_mean - (beta_1*x_train_mean)
print(beta_1)
print(beta_0)
```

```python
y_pred = []
for i in range(sz_test_data):
    y_pred.append( beta_0 + (beta_1*x_test[i]) )
```

```python
y_pred
```

```python
MSE = 0.0 
for y, yp in zip(y_test, y_pred):
    MSE += (yp - y)**2.0

MSE = MSE / (2*sz_test_data)
print(MSE)
```
